## Seaborn Library!

```python
import seaborn as sns

sns.get_dataset_names()
# Arguments x, y, size, and hue are set equal to a column. 
# palette argument requires hue argument.
```

# Scatter Plots

```python
flights = sns.load_dataset("flights")
sns.scatterplot(data = flights, x='year', y='passengers',
                hue = 'month', palette='YlGnBu')
#flights
```

# Histograms and Kernel Density Estimate Plots (KDEs)

```python
titanic = sns.load_dataset('titanic')
sns.histplot(titanic['age'], kde = True)
#titanic 
```

# Bar Plots

```python
tips = sns.load_dataset('tips')
sns.barplot(data = tips, x = 'sex', y = 'tip', hue = 'sex', palette = 'YlGnBu')
#tips
```

# Box Plots

```python
titanic = sns.load_dataset('titanic')
sns.boxplot(data = titanic, x = 'sex', y = 'age')
#titanic
```
